### 家庭
| 欄位內容 | 欄位名稱 |
|---|---|
| 教育程度 | w4p104, 117 |
| 教養態度 | w4p216 - 229 |
| 經濟狀況 | |
| 夫妻關係 | w4p103, 114 |
| 親子關係 | w4p102, 113 |
| 性別 | w4p101 |
| 生活習慣 | w4p118 - 123 |

```python
educationLevel = parent_survey[["w4p104", "w4p117"]]
parentingStyle = parent_survey[[f"w4p{i}" for i in range(216, 230) if i != 217 and i != 218]]  # w4p217 & w4p218 not in columns
parentRelation = parent_survey[["w4p103", "w4p114"]]
parent_childRelation = parent_survey[["w4p102", "w4p113"]]
parentGender = parent_survey[["w4p101"]]
lifeStyle = parent_survey[[f"w4p{i}" for i in range(118, 124)]]

# 有許多NaN需處理
familyFactor = pd.concat([educationLevel, parentingStyle, parentRelation, parent_childRelation, parentGender, parentingStyle])
familyFactor
```

### 升學準備
| 欄位內容 | 欄位名稱 |
|---|---|
| 升學準備 | w4s104 - 107, 116 |
| 升學策略（複選題）：上高三以來，除補習之外，你採取了下列哪些策略來準備升學? | <br>w4s1281 - 1285: 策略種類</br><br>w4s1286: 無</br> |
| 升學方向（複選題）：你打算參加哪些類組的考試? | <br>w4s1301 - 1303: 類組</br><br>w4s1304 - 1305: 專、職</br><br>w4s1306: 不參加</br>

```python
temp = [f"w4s{i}" for i in range(104, 108)]
temp.append("w4s116")
testPreparation = student_survey[temp]

prepareStrategy = student_survey[[f"w4s{i}" for i in range(1281, 1286)]]
testTarget = student_survey[[f"w4s{i}" for i in range(1301, 1306)]]

# 有許多NaN需處理
test_preparationFactor = pd.concat([testPreparation, prepareStrategy, testPreparation])
test_preparationFactor
```

### 分析能力表現
| 欄位內容 | 欄位名稱 |

| 「綜合分析能力測驗」中答對的題數 | w4nright |

```python
comprehensiveAnalysis = student_survey[["w4nright"]]
commonAnalysis = student_survey[["w4cfree"]]
mathAnalysis = student_survey[["w4math"]]

# 有許多NaN需處理
analysisFactor = pd.concat([comprehensiveAnalysis, commonAnalysis, mathAnalysis])
analysisFactor
```
