```python
student_survey = pd.read_csv("TEPS 2007 第四波 高中職:五專學生問卷.csv")
parent_survey = pd.read_csv("TEPS 2007 第四波 高中職:五專家長問卷.csv")
```

### 升學準備
| 欄位內容 | 欄位名稱 |
|---|---|
| 升學準備 | w4s104 - 107, 116 |
| 升學策略（複選題）：上高三以來，除補習之外，你採取了下列哪些策略來準備升學? | <br>w4s1281 - 1285: 策略種類</br><br>w4s1286: 無</br> |
| 升學方向（複選題）：你打算參加哪些類組的考試? | <br>w4s1301 - 1303: 類組</br><br>w4s1304 - 1305: 專、職</br><br>w4s1306: 不參加</br>

```python
temp = [f"w4s{i}" for i in range(104, 108)]
temp.append("w4s116")
testPreparation = student_survey[temp]

prepareStrategy = student_survey[[f"w4s{i}" for i in range(1281, 1286)]]
testTarget = student_survey[[f"w4s{i}" for i in range(1301, 1307)]]

test_preparationFactor = pd.concat([testPreparation, prepareStrategy, testTarget], axis=1)
test_preparationFactor.insert(0, "stud_id", student_survey["stud_id"])
test_preparationFactor["w4s106"] = test_preparationFactor["w4s106"].replace(6, 0)  # 校內課輔時數
test_preparationFactor["w4s107"] = test_preparationFactor["w4s107"].replace(6, 0)  # 校外補習時數
# test_preparationFactor = test_preparationFactor[~((test_preparationFactor == 99) | (test_preparationFactor == 97)).any(axis=1)]
test_preparationFactor
```

### 分析能力表現
| 欄位內容 | 欄位名稱 |
|---|---|
| 「綜合分析能力測驗」中答對的題數 | w4nright |
| 「一般分析能力測驗」中答對的題數 | w4cfree |
| 「數學分析能力測驗」中答對的題數 | w4math |

```python
comprehensiveAnalysis = student_survey[["w4nright"]]
commonAnalysis = student_survey[["w4cfree"]]
mathAnalysis = student_survey[["w4math"]]

analysisFactor = pd.concat([comprehensiveAnalysis, commonAnalysis, mathAnalysis], axis=1)
analysisFactor.insert(0, "stud_id", student_survey["stud_id"])
analysisFactor = analysisFactor.dropna()
analysisFactor['w4nright'] = analysisFactor['w4nright'].replace(' ', np.nan)
analysisFactor['w4cfree'] = analysisFactor['w4cfree'].replace(' ', np.nan)
analysisFactor['w4math'] = analysisFactor['w4math'].replace(' ', np.nan)
analysisFactor['w4nright'] = analysisFactor['w4nright'].fillna(0).astype(int)
analysisFactor['w4cfree'] = analysisFactor['w4cfree'].fillna(0).astype(int)
analysisFactor['w4math'] = analysisFactor['w4math'].fillna(0).astype(int)

analysisFactor
```

```python
# 包含家長教育程度 => Pseudo R-squ.: 0.1491
reg2_1 = pd.merge(test_preparationFactor, educationLevel[["stud_id", "w4p104"]], on="stud_id")

conditions = [
    reg2_1["w4s1301"] == 1,
    reg2_1["w4s1302"] == 1,
    reg2_1["w4s1303"] == 1,
    reg2_1["w4s1304"] == 1,
    reg2_1["w4s1305"] == 1,
    reg2_1["w4s1306"] == 1
]
choices = [1, 2, 3, 4, 5, 0]
reg2_1["group"] = np.select(conditions, choices, default=None)

reg2_1.drop([f"w4s{i}" for i in range(1301, 1307)], axis=1, inplace=True)
reg2_1 = reg2_1[~((reg2_1 == 99) | (reg2_1 == 97)).any(axis=1)]
reg2_1 = reg2_1.dropna(subset=["group"])

X = reg2_1.drop(["group"], axis=1)
y = reg2_1["group"]
y = y.astype('int')

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2023, stratify=y)

model2_1 = LogisticRegression(random_state=0, multi_class="multinomial", penalty="none", solver="newton-cg").fit(X_train, y_train)
preds = model2_1.predict(X_test)

params = model2_1.get_params()
# print(params)

#Print model parameters
# print('Intercept: \n', model2_1.intercept_)
# print('Coefficients: \n', model2_1.coef_)

#Calculate odds ratio estimates
np.exp(model2_1.coef_)

#Use statsmodels to assess variables
logit_model=sm.MNLogit(y_train,sm.add_constant(X_train))
logit_model
result=logit_model.fit()
stats1=result.summary()
stats2=result.summary2()
print(stats1)
print(stats2)
```
